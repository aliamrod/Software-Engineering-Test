```python
# Build a Python application that uses the USGS GeoJSON Feed to plot the average earthquake magnitude
# over the past 7 days.
```

```python
# importing modules
import pandas as pd
import urllib.request
import json
import matplotlib.pyplot as plt
```

```python
# Pick JSON directly from web utilizing the following code.
with urllib.request.urlopen("https://earthquake.usgs.gov/earthquakes/feed/v1.0/summary/all_week.geojson") as url:
    data = json.loads(url.read().decode())
```

```python
df = pd.DataFrame.from_dict(data['features'])
```

```python
# Creating new df from the properties column as the values of properties column
new_df = pd.DataFrame(df['properties'].values.tolist())
```

```python
new_df # prints data onto notebook
```

```python
new_df.plot(x = "time", y = "mag", figsize = (20,8))
plt.xlabel("Day")
plt.ylabel("Cumulative Magnitude")
plt.title("Average Earthquake Magnitude Over 7 Day Period, [Extracted from: USGS GeoJSON Feed]")


plt.show()
```
